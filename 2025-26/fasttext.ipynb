### Tecnologie dei dati e del linguaggio

```python
import fasttext
import fasttext.util
import numpy as np

# Scarica automaticamente il modello italiano (circa 4.2 GB)
fasttext.util.download_model('it', if_exists='ignore')
ft = fasttext.load_model('cc.it.300.bin')

# Esempio di utilizzo
vector = ft.get_word_vector('cane')
```

### Visualizzazione dei vettori

```python
vector = ft.get_word_vector('piano')
print(vector)
print(vector.shape)
```

### Parole più simili

```python
for sigma, w in ft.get_nearest_neighbors('piano', k=20):
    print(w, np.round(sigma, 3))
```

### Composizionalità

```python
for sigma, w in ft.get_analogies('parigi', 'francia', 'italia'):
    print(w, np.round(sigma, 3))
```

### Altre funzioni utili

- `ft.get_sentence_vector('frase intera')` — vettore medio di una frase
- `ft.get_words()` — lista di tutte le parole nel vocabolario
- `ft.get_word_id('parola')` — indice della parola nel vocab (-1 se assente)
- `ft.words` — attributo diretto con la lista delle parole

- `ft.get_subwords('parola')` — restituisce i subword e i loro indici
- `ft.get_subword_vector('ng')` — vettore di un singolo n-gramma

- `ft.predict('testo', k=3)` — top-k label predette con probabilità
- `ft.test('file.txt')` — precision e recall su un file di test

- `fasttext.util.reduce_model(ft, 100)` — riduce le dimensioni dei vettori (es. da 300 a 100)
- `ft.save_model('output.bin')` — salva il modello
- `ft.get_dimension()` — restituisce il numero di dimensioni
