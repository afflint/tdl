## Testi letterari

```python
ps_file = "./data/fermo-e-lucia.txt"
dc_file = "./data/divina-commedia.txt"
ps_corpus = MarkovLM.read_txt(ps_file)
dc_corpus = MarkovLM.read_txt(dc_file)
```

```python
P = MarkovLM(k=4)
D = MarkovLM(k=4)
R = MarkovLM(k=4)
P.train(corpus=ps_corpus)
D.train(corpus=dc_corpus)
R.train(corpus=rc_corpus)
```

```python
print(P.generate(unigram=True))
print(P.generate(unigram=False))
```

```python
print(D.generate(unigram=True))
print(D.generate(unigram=False))
```

```python
print(R.generate(unigram=True))
print(R.generate(unigram=False))
```

## Bigrams

```python
R2 = MarkovLM(k=2)
R2.train(corpus=rc_corpus)
```

```python
pd.Series(R2.k_index[('caffe',)]).sort_values(ascending=False).head(20)
```

```python
for prefix, data in R2.k_index.items():
    if len(data) > 10:
        print(prefix, dict(data))
```

```python
C = 0
for p, c in R2.k_index.items():
    C += sum(c.values())
print(C)
```
